```python
#Importing pandas,numpy,textdistance libraries
import pandas as pd
import numpy as np
#If you don't have textdistance library installed already 
#Please use "!pip install textdistance" command in your jupyter notebook to install textdistance
import textdistance
import re
```

```python
#Loading the textfile from .txt file
words = []
with open('moby.txt', 'r') as tmp:
    #tmp.read() reads the data from the tmp
    files = tmp.read()
    #files.lower() Lowers the case of the text present
    files=files.lower()
    #findall will find all the occurences
    words = re.findall('\w+',files)
```

```python
# This is our collection of words
#set(words) will eliminate the recurrence of the same word
unique_words= set(words)
print("Some of unique elements are : \n",list(unique_words)[0:5])
```

```python
#Findind the frequency of the words
from collections import Counter
frequency_words = {}  
frequency_words = Counter(words)
#frequency_words.most_common() Return a list of the n most common elements and their counts from the most common to the least
print(frequency_words.most_common()[0:10])
```

```python
#finding the probability of the frequency of words
probability_of_words = {}     
number_of_total_words = sum(frequency_words.values())    
for word in frequency_words.keys():
    probability_of_words[word] = frequency_words[word]/number_of_total_words
print("These are the Probabilities of each words: \n",probability_of_words)
```

```python
#Defining a function to return 
def autocorrect(word):
    word = word.lower()
    if word in unique_words:
        #If the same word present in the list it return this
        return('Your word seems to be correct')
    else:
        similar=[]
        for v in frequency_words.keys():
          similar.append(1-(textdistance.Jaccard(qval=2).distance(v,word)))
        #defining df as a Dataframe
        df = pd.DataFrame.from_dict(probability_of_words, orient='index').reset_index()
        #renaming the column names
        df = df.rename(columns={'index':'Word', 0:'Probabilities'})
        #setting the similar values to the similarity column
        df['Similarity'] = similar
        #sorting the values in descending order
        output = df.sort_values(['Similarity', 'Probabilities'], ascending=False).head()
        return(output)
```

```python
#printing the relevant words with correct spelling.
print(list(autocorrect('howeveror')["Word"]))
print(autocorrect('howeveror'))
```
